```python
import sys
import subprocess
import pkg_resources

required = {'autoviz'}
installed = {pkg.key for pkg in pkg_resources.working_set}    # set of all installed packages in the current python environment
missing = required - installed

if missing:
    python = sys.executable
    subprocess.check_call([python, '-m', 'pip', 'install', *missing], stdout=subprocess.DEVNULL)

import pandas as pd
import numpy as np
from autoviz import AutoViz_Class
import ssl

ssl._create_default_https_context = ssl._create_unverified_context
```

```python
AV = AutoViz_Class()
```

```python
path = ''
filename = 'https://raw.githubusercontent.com/harmanani/AAI614/main/Week%202/Boston.csv'
sep = ','
target = 'chas'
```

```python
df = pd.read_csv(path+filename,sep=sep)
dft = df[:]    # create copy
print(df.shape)
df.head(1)
```

```python
dft = AV.AutoViz(path+filename, sep, depVar=target, dfte="",
                header=0, verbose=1,
                lowess=False,chart_format='bokeh',max_rows_analyzed=150000,max_cols_analyzed=30)
```
